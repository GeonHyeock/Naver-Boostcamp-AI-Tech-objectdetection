```python
import glob
import random
import shutil
import os
```

```python
PATH  = "/opt/ml/dataset/yolo_train1/"

filelist = glob.glob(PATH + '*.txt')
filelist
# filelist[0]
# filelist[0][-8:-4]
```

```python
filelist[0][-8:-4]
```

```python
namelist = [i[-8:-4] for i in filelist]
namelist.sort()
namelist
```

```python
random.seed(2022)
random.shuffle(namelist)
```

```python
data_size = len(namelist)
num_val1 = int(data_size * 0.2)
num_val2 = num_val1 * 2
num_val3 = num_val1 * 3
num_val4 = num_val1 * 4
```

```python
name_val1, name_train1 = namelist[:num_val1], namelist[num_val1:]
name_val2, name_train2 = namelist[num_val1:num_val2], namelist[:num_val1] + namelist[num_val2:]
name_val3, name_train3 = namelist[num_val2:num_val3], namelist[:num_val2] + namelist[num_val3:]
name_val4, name_train4 = namelist[num_val3:num_val4], namelist[:num_val3] + namelist[num_val4:]
name_val5, name_train5 = namelist[num_val4:], namelist[:num_val4]
```

```python
name_train2
```

```python
name_val2
```

```python
val_list = [name_val1, name_val2, name_val3, name_val4, name_val5]
train_list = [name_train1, name_train2, name_train3, name_train4, name_train5]
num_list = [1, 2, 3, 4, 5]
```

```python
for num, val in zip(num_list, val_list):
    print(num, val)
    break
```

```python
output_path = '/opt/ml/dataset/yolo_train/val'

for num in num_list:
    print('/opt/ml/dataset/yolo_train/val' + str(num))
```

```python
output_path = '/opt/ml/dataset/yolo_train1/val'

for num, val in zip(num_list, val_list):
    output_path = '/opt/ml/dataset/yolo_train1/val' + str(num) + '/'
    for name in val:
        txt_path = os.path.join(PATH + name + '.txt')
        img_path = os.path.join(PATH + name + '.jpg')
        print(txt_path, img_path, output_path)
        
        shutil.copy(txt_path, output_path)
        shutil.copy(img_path, output_path)
```

```python
output_path = '/opt/ml/dataset/yolo_train1/train'

for num, train in zip(num_list, train_list):
    output_path = '/opt/ml/dataset/yolo_train1/train' + str(num)
    for name in train:
        txt_path = os.path.join(PATH + name + '.txt')
        img_path = os.path.join(PATH + name + '.jpg')
        print(txt_path, img_path, output_path)
        
        shutil.copy(txt_path, output_path)
        shutil.copy(img_path, output_path)
```
